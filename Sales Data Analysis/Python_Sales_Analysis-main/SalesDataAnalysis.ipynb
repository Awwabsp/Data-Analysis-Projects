```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv", encoding='latin1')
df
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"], axis=1 , inplace=True)
df
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df["Amount"] = df["Amount"].astype("int")
df["Amount"].dtypes
```

```python
ANALYSIS
```

```python
gp = sns.countplot(x = "Gender",data = df, hue="Gender")
for bars in gp.containers:
    gp.bar_label(bars)
plt.show()
```

```python
Gender_amount = df.groupby(["Gender"])["Amount"].sum()
ga = pd.DataFrame(Gender_amount)
```

```python
sns.barplot(x="Gender", y="Amount", data= ga,hue="Gender")
plt.show()
```

**from the above graphs we can conclude that most of the buyers are females and the purchasing power of females is greater than man**

```python
ag = sns.countplot(x="Age Group" , data = df, hue = "Gender")
for bars in ag.containers:
    ag.bar_label(bars)
plt.show()
```

```python
sales_age = df.groupby(["Age Group"])["Amount"].sum()
sa = pd.DataFrame(sales_age)
sns.barplot(x = "Age Group", y = "Amount"  , data = sa , hue = "Age Group" , palette = "magma")
```

**from the above graph we conclude that most buyers are females belonging to the age group 26-35**

```python
states_order = df.groupby(["State"])["Orders"].sum()
so = pd.DataFrame(states_order)
plt.figure(figsize=(25,10))
sns.barplot(x = "State", y="Orders", data = so, hue="State", palette="viridis")
plt.show()
```

**from the above graph we conclude that most orders are from Uttar Pradesh , Maharashtra and Kamataka**

```python
ms = sns.countplot(x="Marital_Status" , data = df , hue = "Gender", palette="viridis")
for bars in ms.containers:
    ms.bar_label(bars)
```

**from above graph we conclude that most of the buyers are married females**

```python

plt.figure(figsize=(18,10))
occ=sns.countplot(x="Occupation", data=df , hue="Gender")
for bars in occ.containers:
    occ.bar_label(bars)
plt.show()
```

```python

plt.figure(figsize=(18,10))
occ_amount = df.groupby(["Occupation"])["Amount"].sum().sort_values(ascending=False)
occ_a = pd.DataFrame(occ_amount)
occ=sns.barplot(x="Occupation", y="Amount", data=occ_a, hue="Occupation", palette="viridis")
plt.show()
```

**from above graphs we conclude that most buyers belong to IT Sector , HealthCare and Aviation respectivley**

```python
plt.figure(figsize=(25,8))
sns.countplot(x="Product_Category", data=df , hue="Product_Category", palette="viridis")
plt.show()
```

<span style="font-size: 30px; font-weight:bolder;">Conclusion:</span>


**Married women belonging to the age group 26-35 yrs from UP, Maharastra and kamataka working in IT, Healthcare and Aviation are more likely to buy from food , clothing and electronics category**
